```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../data/netflix_titles.csv')
```

```python
# 1. Movies vs. TV Shows
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Movies vs. TV Shows')
plt.show()
```

```python
# 2. Top Genres
df['listed_in'] = df['listed_in'].apply(lambda x: x.split(', '))
genres = df.explode('listed_in')['listed_in']
plt.figure(figsize=(10, 8))
genres.value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Genres')
plt.show()
```

```python
# 3. Content Over the Years
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
plt.figure(figsize=(12, 6))
df['release_year'].value_counts().sort_index().plot(kind='line', color='green')
plt.title('Content Over the Years')
plt.show()
```

```python
# 4. Top Countries
df['country'] = df['country'].apply(lambda x: x.split(', ')[0] if isinstance(x, str) else 'Unknown')
plt.figure(figsize=(10, 8))
df['country'].value_counts().head(10).plot(kind='barh', color='orange')
plt.title('Top 10 Countries')
plt.show()
```
